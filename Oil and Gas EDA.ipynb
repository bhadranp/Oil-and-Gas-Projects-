```python
plt.figure(figsize=(20,5))
sns.lineplot(data=oil,x='Depth',y='RxoRt')
plt.title('Rxo / Rt',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='RLL3',color='red')
plt.title('Shallow Resistivity',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='SP',color='darkred')
plt.title('Spontaneous Potential',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='RILD',color='indigo')
plt.title(' Deep Resistivity',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='MN',color='crimson')
plt.title('Micro Normal',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='MI',color='darkslategrey')
plt.title('Micro Inverse',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='MCAL',color='midnightblue')
plt.title('Micro Caliper',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='DCAL',color='darkmagenta')
plt.title('Caliper',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='RHOB',color='black')
plt.title('Bulk Density',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='RHOC',color='darkorange')
plt.title('Density Correction',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='DPOR',color='blue')
plt.title('Density Porosity',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='CNLS',color='deeppink')
plt.title('Limestone Porosity',fontsize='40')
plt.show()

plt.figure(figsize=(40,9))
sns.lineplot(data=oil,x='Depth',y='GR',color='darkolivegreen')
plt.title('Gamma Ray',fontsize='40')
plt.show()
```
